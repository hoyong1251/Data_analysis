```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset('planets')
df.head()
```

```python
Q25 = df['orbital_period'].quantile(0.25)
Q75 = df['orbital_period'].quantile(0.75)
IQR = Q75-Q25
min = Q25 - IQR * 1.5
max = Q75 + IQR * 1.5
df_outlier = df[(df['orbital_period']<=min) | (df['orbital_period']>=max)]
df_outlier
```
