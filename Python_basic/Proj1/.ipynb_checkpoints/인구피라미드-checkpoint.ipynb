## 남자 데이터 정의

```python
import pandas as pd
file_name = '202205_202205_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name,skiprows=3,index_col='행정기관',usecols='B,E:Y')
df_m
```

```python
# 맨위 전국 데이터 로우만 볼건데 숫자에 콤마때문에 문자열 인식할수도 있음
# 이를 숫자로 바꾸는 작업

df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
```

```python
df_m.iloc[0]
```

## 여자데이터 정의

```python
df_w = pd.read_excel(file_name,skiprows=3,index_col='행정기관',usecols='B,AB:AV')
df_w
```

```python
## 여자데이터 col 이름이 조금씩 다른데 앞에 남자데이터와 동일해서 
# 구분하기 위함임 / 이를 바꿔줘야함

df_w.columns = df_m.columns
df_w.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w
```

## 그래프 그리기

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns ,- df_m.iloc[0] // 1000)  #단위 :천명 / 남자데이터를 음수로 해서 반대방향 으로 막대 지정
plt.barh(df_w.columns, df_m.iloc[0] // 1000)
plt.title("2022년 5월 대한민국 인구 피라미드")
plt.savefig("2022_인구 피라미드.png",dpi=100)
plt.show()
```
